 ### Get definition link for each word (In French)

```python

import requests

# The following is neccesary because the URLs of the dictionary entries used cannot be easily deduced from the words themselves.
# For example, the word "Bagguete" corresponds to the following URL: https://www.dictionnaire-academie.fr/article/A9B0115"

endpoint = "https://www.dictionnaire-academie.fr/search"
headers = {
    "Accept": "application/json, text/javascript, */*; q=0.01",
    "Content-Type": "application/x-www-form-urlencoded; charset=UTF-8",
    "X-Requested-With": "XMLHttpRequest",
    "Origin": "https://www.dictionnaire-academie.fr",
    "Referer": "https://www.dictionnaire-academie.fr/",
    "User-Agent": "Mozilla/5.0"
}

def word_def_links(word):
    payload = {"term": word}
    r = requests.post(endpoint, data=payload, headers=headers)
    response_json = r.json()
    df = pd.DataFrame(response_json['result'])

    
    
    # We get a dictionary as an answer, some of its fields are:
    # url: URL assigned to a word (One URL for each homograph).
    # label: Searched word.
    # score: It ranks, from 0 to 1, the similarity of the lab to a word entry.
    # nbhomograph: Ranks, from more popular to least, meaning of words with various meanings but same writing (homographs).

    # I'll assume the frequency list is perfectly written. So, I'll discard anything other than a perfect 1 score.
    df_perfect = df[df['score'] == '1.0']

    # If a word has an homograph, multiple rows result. Which URL should I choose?
    
    # Homographs are important and they're not contemplated for in the frequency list used. Therefore, I made the
    # decision to keep all url's and give the user control over what they think is important.
    
    return df_perfect

word_def_links("porter")
```

```python
def process_row(row):
    freq = row["Frequency"]
    word = row["Lemme"]
    def_links_results = word_def_links(word)
    sound_link_result = f"https://forvo.com/word/{word}"

    if def_links_results.empty:
        print(f"{word} has not been found in dictionary.")
        return


    def_links_results.drop(columns=["score"], inplace=True)


    # Create a temp DataFrame

    temp = def_links_results.copy()
    temp['Frequency'] = freq
    temp['sound_link'] = f"https://forvo.com/word/{word}"
    
    # Mark if this word has multiple results
    temp['multiple_results'] = len(temp) > 1
    
    expanded_rows.append(temp)

    if current_index % 10 == 0:
        write_results(expanded_rows)
        expanded_rows.clear()
    
    return current_index + 1
    
    
```

```python
def process_words_parallel(df_words, csv_filepath):
    not_found = []
    
    # Check if file exists, if not, generate header.
    if not os.path.exists(csv_filepath):
        header = ['Frequency', 'Lemma', 'Grammatical Function', 'Homograph Number', 'Definition URL', 'Sound URL', 'Homograph']
        df_header = pd.DataFrame(columns=header)
        df_header.to_csv(csv_filepath, mode='w', header=True, index=False)
        

    with concurrent.futures.ThreadPoolExecutor() as executor:
        futures = []  # Store future results
        
        for _, row in df_words.iloc[:100].iterrows():  
            freq = row["Frequency"]
            word = row["Lemme"]
            
            # Submit each word processing task to the executor
            futures.append(executor.submit(process_row, row))

        # Wait for all threads to complete and collect the results
        for future in concurrent.futures.as_completed(futures):
            result = future.result()
            if result is not None:
                current_index = result  # Update the index from the result

    # Save any remaining results
    if expanded_rows:
        write_results(expanded_rows)
    
    return not_found
```

### What now?

I'll create a program that receives an URL that includes a timestamp, this is easily generated by right clicking any youtube video and selecting "Copy video URL at current time".

This is the result: **https://youtu.be/DrtP0F-dcXk?t=59**

It contains everything we need to download a short clip, along with the official or autogenerated transcription.


```python
from youtube_transcript_api import YouTubeTranscriptApi

ytt_api = YouTubeTranscriptApi()
```

```python
def fetch_transcript(video_id):
    try:
        transcript = ytt_api.fetch(video_id)
        return transcript
    except Exception as e:
        print(f"Error fetching transcript: {e}")
        return None
```

```python
import re
```

```python
# From the URL we can extract the time in seconds
URL = "https://youtu.be/DrtP0F-dcXk?t=59"

match = re.search(r'[?&]t=(\d+)', URL)

if match:
    time_in_seconds = int(match.group(1))

time_in_seconds
```

```python
# We only have this methods at our disposal,
dir(test_transcript[0])[-4:]
```

```python
# so here's the plan: By fault, we'll get the transcript that starts just before our selected time.
# let's implement a basic binary search for this purpose.

left = 0
right = len(test_transcript) - 1

best_index = None

while left <= right:
    middle = (left + right)//2

    if test_transcript[middle].start >= time_in_seconds:
        right = middle - 1
        best_index = middle
        
    else:
        left = middle + 1

test_transcript[best_index]    
    
```

### Why saving the index? 
That way, if the transcription is incomplete, we know where to look for the full phrase in the transcriptions list.

```python

print("Current transcription window: ", [best_index, best_index])
decision = input("Change [Y/N]")
if decision.lower != "n":
    left = int(input("New LEFT index: "))
    right = int(input("New RIGHT index: "))

    print("UPDATED: ", (left, right))
    
for transcript in test_transcript[left:right]:
    print(transcript.text)
    
```
